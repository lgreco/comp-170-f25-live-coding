```python
def is_greater(a, b):
    test_a_more_than_b = a > b
    if test_a_more_than_b:
        print(f"{a} is greater than {b}")
    else:
        print(f"{a} is less than or equal to {b}")
```

```python
is_greater(100, 20)
```

```python
def string_greater(a:str, b:str) -> bool:
    # assume that a is greater
    result = False
    # start an index to go through the strings
    i = 0
    while i < len(a) and i < len(b) and not result:
        result = ord(a[i]) > ord(b[i])
        i = i+1
    return result
```

```python
string_greater("airport", "airplane")
```
