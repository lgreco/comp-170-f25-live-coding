```python
# Shallow copy .. fertile ground for bugs 
true_LOTR_heroes = ['Gandalf', 'Legolas', 'Frodo']
wannabes = true_LOTR_heroes
wannabes.append('Leo')
print(true_LOTR_heroes)
```

```python
# Pythonic deep copy
true_LOTR_heroes = ['Gandalf', 'Legolas', 'Frodo']
wannabes = [name for name in true_LOTR_heroes] # list comprehension
wannabes.append('Leo')
print(true_LOTR_heroes)
```

```python
# List comprehension, step by step
true_LOTR_heroes = ['Gandalf', 'Legolas', 'Frodo']  # Initial list
wannabes = []  # initialize the list that will become the deep copy
for i in range(len(true_LOTR_heroes)):  # iterate over original list
    wannabes.append(true_LOTR_heroes[i])  # add a copy of an original list element to copy list
print(f"{wannabes=}")
wannabes.append('Leo')
print(f"{true_LOTR_heroes=}")
print(f"{wannabes=}")
```

```python
# Deep copy by treating lists as arrays
true_LOTR_heroes = ['Gandalf', 'Legolas', 'Frodo']  # Initial list
# Create a list that will become the deep copy and that will have
# as many elements as the input list.
wannabes = [None] * len(true_LOTR_heroes)
print(f"{wannabes=}")
for i in range(len(true_LOTR_heroes)):
    wannabes[i] = true_LOTR_heroes[i]

print(f"{wannabes=}")
wannabes.append('leo')
print(f"{wannabes=}")
print(f"{true_LOTR_heroes=}")
```

```python
# Standard Python way
import copy
true_LOTR_heroes = ['Gandalf', 'Legolas', 'Frodo']  # Initial list
wannabes = copy.deepcopy(true_LOTR_heroes)
wannabes.append("leo")
print(f"{true_LOTR_heroes=}")
```
