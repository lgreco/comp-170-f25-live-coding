# Exam 2 practice

Write a method

```python
def count_words_of(length:int, string:str) -> int:
```

to accept an `int` value and a `string` and count how many words in the string have the length specified by the integer number. For example,

```python
def count_words_of(3, "now is the winter of our discontent")
```

should return `3` because the string contains 3 words of length 3.

The values for `length` and `string` should be obtained from the keyboard and validated before passed to the method. Use a separate method to report (`print`) the outcome of `count_words_of`.

You may need to write additional methods at your discretion.


````python
def obtain_data():
    """Obtain and validate data for the method count_words_of.
    This method ensures that the returned integer >= 0 and that
    the input string is not None. Empty strings are acceptable.
    """
    string = input("Enter string to process: ")
    if string is None:
        raise ValueError("String must not be null")
    length = input("Enter length to search for: ")
    if not length.isnumeric():
        raise ValueError("length must be an integer number")
    length = int(length)
    if length < 0:
        raise ValueError("length must be non negative")
    # Clean data, call the method:
    answer = count_words_of(length, string)
    print(f"There are {answer} token(s) of length {length} in the following string:\n{string}")

def count_words_of(length: int, string: str) -> int:
    """Finds how many tokens of the specified length are 
    present in the input string. The method assumes that
    both its arguments are valid valued.
    """
    # Obtain list of tokens in the string
    tokens = string.split()
    count = 0
    # Consider every element in list of tokens
    for token in tokens:
        if len(token) == length:
            # Update count each time we find a token of the
            # specified length
            count += 1
    return count
````
